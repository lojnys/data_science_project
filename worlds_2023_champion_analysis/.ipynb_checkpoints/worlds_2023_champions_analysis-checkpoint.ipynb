The game is organized in a SWISS format. The overall idea of SWISS stage is that teams that reach 3 wins within the series advance to Knockout stage and for those we have 3 losses within the series are eliminated. 
* Teams first randomly go up against each other and then there will be two brackets: winners' bracket and losers' bracket. 
* After this round, teams will compete each other in each bracket, then proceed to the next round with 3 sub-groups: a group of teams with score of 2:0, a group of teams with score of 1:1, and a group of teams with schore of 0:2.
* These rounds keeps on going untill there are teams with 3 losses and teams with 3 wins. 

Total of 8 teams will advance to the Knockout stage. In the Knowckaout stage, it is a tournament format, where if a team loses, they are immediately eliminated from the series.

## Champion Rankings by Pick/Ban Rate and Winrate

In this section, we will analyze the pick/ban rate and winrate to decide which the rankings for each position (top, mid, jungle, adc, support). The criterion to be considered as a top tier champions are:

* The winrate must be greater than equal to 50% and
* The pick/ban rate must be greater than eqal to 30% and
* Each champion should be used by multiple players *

*: this is because if one player uses the champion and obtain high win rate, then this just means that the player is good at the champion. However, the champion is used by multiple players, then it can be certain that the champion plays the key role in win rate.

```r
library(repr)
library(tidyverse)
library(tidymodels)
```

```r
total_games <- 25
data <- read_csv('data/worlds_2023_champions_stat.csv')
head(data)
```

```r
data_with_rate <- data |>
mutate(win_rate = win / picked) |>
mutate(pb_rate = (picked + banned)/total_games) |>
mutate(pick_rate = picked / total_games) |>
mutate(ban_rate = banned / total_games)
head(data_with_rate)
```

```r
options(repr.plot.width = 20, repr.plot.height =10) 

champion_rate_plot <- data_with_rate |>
ggplot(aes(x = win_rate, y = pb_rate, color = position)) +
geom_point(size = 5) +
labs(x = "Win Rate (%)", y = "Pick/Ban Rate (%)", color = "Position") +
geom_vline(xintercept = 0.5, linetype = "dashed") +
geom_hline(yintercept = 0.3, linetype = "dashed") +
ggtitle("Pick/Ban Rate vs. Win Rate by position") +
facet_grid(cols = vars(position)) +
theme(text = element_text(size = 30), axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1)) 

champion_rate_plot
```

**Analysis** 

The plot above shows the champions which have pick/ban greater than 30% and at the same time, win rate of 50%. Thus the region that contains all the "meta" champions are the right corner of the threshholds. The number of so-called "meta" champions for each position differ by position. For example, high pick/ban rate and high win rate in adc shows only one champion. This means that the champion is the "hot potato" in the game where the players either pick or ban in almost every game. In the top left corner of the thresholds, we have champions who have high win rate but have low pick/ban rate. Although they don't qualify to be the "meta" champions, they could have potential being the joker pick in the game, due to high win rate.

```r
by_vs_winrate_plot <- data_with_rate |>
ggplot(aes(x = By, y = win_rate, color = position)) +
geom_point(size = 5) +
labs(x = "Number of players played the champion", y = "Win Rate (%)", color = "Position") +
ggtitle("By vs. Win Rate by position") +
geom_hline(yintercept = 0.5, linetype = "dashed") +
geom_vline(xintercept = 5, linetype = "dashed") +
facet_grid(cols = vars(position)) +
theme(text = element_text(size = 30))

by_vs_winrate_plot
```

**Analysis**

As mentioned above, high number of players used the champions, correlating with high win rate represents that those champions are popular. This plot won't be the definite way to determine which champions are "meta". However, this information will be the additional evidence to determine the "meta" champions. 

**high pick/ban rate, high win rate & high number of players used, high win rate => meta/popular champion
low pick/ban rate, high win rate & low number of players used, high win rate => high potential/unpopular champion**
